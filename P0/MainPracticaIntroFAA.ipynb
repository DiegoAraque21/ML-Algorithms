```python
from Datos import Datos
import EstrategiaParticionado

d1=Datos('./datasets/heart.csv')
d2=Datos('./datasets/tic-tac-toe.csv')
```

## Dataset Heart 

```python
print("NOMINAL ATRIBUTOS")
print(d1.nominalAtributos)

print("DICCIONARIO")
print (d1.diccionarios)

print("NUMBER OF ROWS")
print(d1.datos.shape[0])

print("MATRIZ DE DATOS")
d1.datos.head()
```

```python
estrategiaVC=EstrategiaParticionado.ValidacionCruzada(10)
estrategiaVC.creaParticiones(d1.datos)       # datos se refiere a la matriz numérica calculada en Datos
print("VALIDACIÓN CRUZADA")
print("Train-particion 0: ", estrategiaVC.particiones[0].indicesTrain)       # particiones contiene la lista de particiones
print("Test-particion 0: ", estrategiaVC.particiones[0].indicesTest)

print("Número de indices de train: ", len(estrategiaVC.particiones[0].indicesTrain))
print("Número de indices de test: ", len(estrategiaVC.particiones[0].indicesTest))
```

```python
estrategiaVS=EstrategiaParticionado.ValidacionSimple(2, 0.3)
estrategiaVS.creaParticiones(d1.datos)       # datos se refiere a la matriz numérica calculada en Datos
print("VALIDACIÓN SIMPLE")
print("Train-particion 0: ", estrategiaVS.particiones[0].indicesTrain)       # particiones contiene la lista de particiones
print("Test-particion 0: ", estrategiaVS.particiones[0].indicesTest)

print("Número de indices de train: ", len(estrategiaVS.particiones[0].indicesTrain))
print("Número de indices de test: ", len(estrategiaVS.particiones[0].indicesTest))
```

```python
# utilizar el método extraeDatos para obtener los datos de entrenamiento y test para cada partición
print("DATOS DE ENTRENAMIENTO")
d1.extraeDatos(estrategiaVC.particiones[0].indicesTrain).head()
```

```python
print("DATOS DE TEST")
d2.extraeDatos(estrategiaVC.particiones[0].indicesTest).head()
```
